```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("EmployeeProcessing") \
    .getOrCreate()
```

```python
df = spark.read.csv("employees.csv", header=True, inferSchema=True)
df.show()
```

```python
# 1. Sort Employees by Salary (Descending)
sorted_df = df.orderBy(df.salary.desc())
sorted_df.show()
```

```python
#2. Department-wise Total Salary
from pyspark.sql.functions import sum

dept_salary = df.groupBy("department") \
                .agg(sum("salary").alias("total_salary"))
dept_salary.show()
```

```python
# 3. Top 3 Highest Paid Employees
top3 = df.orderBy(df.salary.desc()).limit(3)
top3.show()
```

```python
# 4. Save Output
with open("output.txt", "w") as f:

    f.write("Employees Sorted by Salary (Descending)\n")
    f.write(sorted_df.toPandas().to_string())

    f.write("\n\nDepartment-wise Total Salary\n")
    f.write(dept_salary.toPandas().to_string())
    
    f.write("\n\nTop 3 Highest Paid Employees\n")
    f.write(top3.toPandas().to_string())

print("Output saved successfully")
```

```python
spark.stop()
```

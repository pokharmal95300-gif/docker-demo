```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
spark = SparkSession.builder \
    .appName("Sales Analysis") \
    .getOrCreate()

print("Spark Session Created Successfully")
```

```python
df = spark.read.csv(
    "sales_data.csv",
    header=True,
    inferSchema=True
)

print("Dataset Loaded Successfully")
df.show()
```

```python
df.printSchema()
```

```python
# 1. Sort Products by Sales (Descending Order)
print("Products Sorted by Sales (Descending Order)")
sorted_df = df.orderBy(col("sales").desc())
sorted_df.show()
```

```python
# 2. Top 3 Highest-Selling Products
print("Top 3 Highest-Selling Products")
top3_df = sorted_df.limit(3)
top3_df.show()
```

```python
# 3. Filter Products with Sales Greater Than 80,000
print("Products with Sales Greater Than 80,000")
filtered_df = df.filter(col("sales") > 80000)
filtered_df.show()
```

```python
# 4. Save Filtered Results to CSV
filtered_df.write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv("filtered_output")

print("Filtered data saved successfully.")
```

```python
# Count Total Products
print("Total Number of Products:")
df.count()
```

```python
spark.stop()
print("Spark Session Stopped")
```
